## Opening netcdf4

**Idea**: xarray can get very slow to open a netcdf file, especially if it is a large one. This is not necessary to get some simple information such as the first year of the simulation. Using netCDF4 package?

*example* getting the first year of a netCDF time series with xarray:

```python
start = time.time()
ts = xr.open_dataset("/nfs/see-fs-01_users/eeymr/database/xoupb/time_series/xoupb.oceansalipg16.annual.nc")
print(ts.t.values[0].year)
print(f"Xarray : {time.time() - start}")
```

```python
start = time.time()
times = netCDF4.Dataset("/nfs/see-fs-01_users/eeymr/database/xoupb/time_series/xoupb.oceansalipg16.annual.nc").variables['t']
print(netCDF4.num2date(times[:],units=times.units,calendar=times.calendar)[0].year)
print(f"Xarray : {time.time() - start}")
```
